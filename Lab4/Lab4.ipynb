#### Thao tác lọc dữ liệu filter:
1. Lấy dữ liệu về số ca nhiễm ở Trung Quốc (Mainland China):

```python
df_covid[df_covid["Country/Region"] == "Mainland China"]
```

2. Lấy dữ liệu về số ca nhiễm của Việt Nam trong tháng 03 và tháng 04.

```python
df_covid[(df_covid["Country/Region"] == "Vietnam") 
                         & (df_covid["ObservationDate"] >= "2020-03-01") 
                         & (df_covid["ObservationDate"] <= "2020-04-30")]
```

#### Thao tác thống kê dữ liệu với summarise:

Thống kê dữ liệu về số ca dương tính của China: số ca dương tính trung bình, trung vị,
phương sai và độ lệch chuẩn.

```python
df_covid[df_covid["Country/Region"] == "Mainland China"].agg(
    Mean=("Confirmed", "mean"),
    Median=("Confirmed", "median"),
    Variance=("Confirmed", "var"),
    SD=("Confirmed", "std")
)
```

#### Thao tác gom nhóm dữ liệu với group_by:
Hiển thị dữ liệu theo từng ngày quan sát (thuộc tính ObservationDate) của Việt Nam
trong 2 tháng: tháng 3 và tháng 4 năm 2020.

```python
df_covid[
    (df_covid["Country/Region"] == "Vietnam") &
    (df_covid["ObservationDate"] >= "2020-03-01") &
    (df_covid["ObservationDate"] <= "2020-04-30")
].groupby("ObservationDate").agg(list)
```

#### Thao tác sắp xếp dữ liệu với arrange:
1. Hiển thị dữ liệu theo từng ngày quan sát (thuộc tính ObservationDate) của Việt
Nam trong 2 tháng: tháng 3 và tháng 4 năm 2020. Sắp xếp theo số ca dương tính tăng
dần.

```python
df_covid[
    (df_covid["Country/Region"] == "Vietnam") &
    (df_covid["ObservationDate"] >= "2020-03-01") &
    (df_covid["ObservationDate"] <= "2020-04-30")
].groupby("ObservationDate").apply(lambda x: x.sort_values("Confirmed")).reset_index(drop=True)
```

2. Tương tự như trên, nhưng sắp xếp số ca dương tính giảm dần. Để sắp xếp giảm dần
một thuộc tính, ta dùng thao tác: desc(<thuộc_tính>) trong hàm arrange.

```python
df_covid[
    (df_covid["Country/Region"] == "Vietnam") &
    (df_covid["ObservationDate"] >= "2020-03-01") &
    (df_covid["ObservationDate"] <= "2020-04-30")
].groupby("ObservationDate").apply(lambda x: x.sort_values("Confirmed", ascending=False)).reset_index(drop=True)
```
